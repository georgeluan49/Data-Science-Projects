# Part A

## Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] 
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1]
n_cols
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

## define regression model

```python
def regression_modelA():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
ModelA = regression_modelA()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
```

```python
X_test.head()
```

```python
ModelA.fit(X_train, y_train, epochs=50)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
yhat = ModelA.predict(X_test)
```

```python
y_real = y_test
```

```python
mean_squared_error(y_real, yhat)
```

## 50 Iterations

```python
MSEs = []
i = 1
while i <= 50:
    pred_train, pred_test, tar_train, tar_test = train_test_split(predictors, target, test_size=0.3)
    ModelA.fit(pred_train, tar_train, epochs=50)
    y_h = ModelA.predict(pred_test)
    MSE = mean_squared_error(tar_test, y_h,squared=True)
    MSEs.append(MSE)
    i += 1
```

```python
MSEs
```

```python
from statistics import mean
```

```python
Mean = mean(MSEs)
print('Mean:',Mean)
STD = np. std(MSEs)
print('Standard Deviation:',STD)
```

# Part B

## Normalization

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
target_norm = (target - target.mean())/target.std()
target_norm.head()
```

```python
ModelB = regression_modelA()
bMSEs = []
i = 1
while i <= 50:
    predb_train, predb_test, tarb_train, tarb_test = train_test_split(predictors_norm, target_norm, test_size=0.3)
    ModelB.fit(predb_train, tarb_train, epochs=50)
    yb_h = ModelB.predict(predb_test)
    bMSE = mean_squared_error(tarb_test, yb_h,squared=True)
    bMSEs.append(bMSE)
    i += 1
print(mean(bMSEs))
print(np. std(bMSEs))
```

# Part C

```python
ModelC = regression_modelA()
cMSEs = []
i = 1
while i <= 50:
    predc_train, predc_test, tarc_train, tarc_test = train_test_split(predictors_norm, target_norm, test_size=0.3)
    ModelC.fit(predc_train, tarc_train, epochs=100)
    yc_h = ModelC.predict(predc_test)
    cMSE = mean_squared_error(tarc_test, yc_h,squared=True)
    cMSEs.append(cMSE)
    i += 1
print(cMSEs)
print('Mean is', mean(cMSEs))
print('Standard Deviation is', np. std(cMSEs))
```
